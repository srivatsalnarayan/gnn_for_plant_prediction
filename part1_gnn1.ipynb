```python
import torch
from torch_geometric.data import Data

# Dummy graph with 3 nodes and 2 edges
edge_index = torch.tensor([[0, 1, 2], [1, 2, 0]], dtype=torch.long)  # Shape [2, num_edges]
x = torch.tensor([[1, 2], [2, 3], [3, 4]], dtype=torch.float)       # Node features

# Create a dummy Data object
data = Data(x=x, edge_index=edge_index)

print(data)
```

```python
import torch
import torch.nn as nn
import torch_geometric.nn as pyg_nn

class GNNModel(nn.Module):
    def __init__(self, num_features, num_classes):
        super(GNNModel, self).__init__()
        self.conv1 = pyg_nn.GCNConv(num_features, 64)
        self.conv2 = pyg_nn.GCNConv(64, 64)
        self.fc = nn.Linear(64, num_classes)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = torch.relu(x)
        x = self.conv2(x, edge_index)
        x = torch.relu(x)
        x = pyg_nn.global_mean_pool(x, torch.zeros(x.size(0), dtype=torch.long))  # Global pooling
        x = self.fc(x)
        return x
```

```python
import torch.optim as optim
from torch_geometric.loader import DataLoader

# Hyperparameters
num_features = 2  # Number of features per node
num_classes = 2   # Number of classes for classification
learning_rate = 0.01
num_epochs = 20

# Initialize model, criterion, and optimizer
model = GNNModel(num_features, num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Dummy DataLoader for illustration
train_loader = DataLoader([data], batch_size=1, shuffle=True)
test_loader = DataLoader([data], batch_size=1)

def train():
    model.train()
    for batch in train_loader:
        optimizer.zero_grad()
        out = model(batch)
        loss = criterion(out, torch.tensor([0]))  # Dummy target
        loss.backward()
        optimizer.step()
        print(f"Training Loss: {loss.item()}")

def evaluate():
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for batch in test_loader:
            out = model(batch)
            _, predicted = torch.max(out, 1)
            total += 1
            correct += (predicted == torch.tensor([0])).sum().item()  # Dummy target
    accuracy = correct / total
    print(f"Accuracy: {accuracy}")

# Run training and evaluation
train()
evaluate()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data, DataLoader

class GNNModel(nn.Module):
    def __init__(self, num_features, num_classes):
        super(GNNModel, self).__init__()
        self.conv1 = GCNConv(num_features, 16)
        self.conv2 = GCNConv(16, num_classes)
    
    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = torch.relu(x)
        x = self.conv2(x, edge_index)
        return x
```

```python
def load_data_from_csv(csv_path):
    # Load data
    df = pd.read_csv(csv_path)
    
    # Convert categorical labels to integers
    df['label'], _ = pd.factorize(df['label'])
    
    # Normalize numerical features (optional)
    features = df[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']].values
    features = (features - features.mean(axis=0)) / features.std(axis=0)
    
    # Convert DataFrame to node features and edge indices
    num_nodes = len(df)
    num_features = features.shape[1]
    num_classes = len(df['label'].unique())
    
    # Create edge index (fully connected graph for simplicity)
    edge_index = torch.combinations(torch.arange(num_nodes), r=2).t().contiguous()
    
    # Convert to PyTorch tensors
    x = torch.tensor(features, dtype=torch.float)
    y = torch.tensor(df['label'].values, dtype=torch.long)
    
    # Create PyG data object
    data = Data(x=x, edge_index=edge_index, y=y)
    
    return data, num_features, num_classes

# Load data
csv_path = "C:/Users/SRIVATSAL NARAYAN/Desktop/cc/Crop_recommendation.csv"
data, num_features, num_classes = load_data_from_csv(csv_path)

# Create DataLoader
train_loader = DataLoader([data], batch_size=1, shuffle=True)
test_loader = DataLoader([data], batch_size=1)
```

```python
def train(model, train_loader, optimizer, criterion):
    model.train()
    for batch in train_loader:
        optimizer.zero_grad()
        out = model(batch)
        loss = criterion(out, batch.y)
        loss.backward()
        optimizer.step()
        print(f"Training Loss: {loss.item()}")

def evaluate(model, test_loader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for batch in test_loader:
            out = model(batch)
            _, predicted = torch.max(out, 1)
            total += batch.y.size(0)
            correct += (predicted == batch.y).sum().item()
    accuracy = correct / total
    print(f"Accuracy: {accuracy}")

# Initialize model, optimizer, and loss function
model = GNNModel(num_features=num_features, num_classes=num_classes)
optimizer = optim.Adam(model.parameters(), lr=0.01)
criterion = nn.CrossEntropyLoss()

# Run training and evaluation
train(model, train_loader, optimizer, criterion)
evaluate(model, test_loader)
```

```python
import pandas as pd
import torch
import networkx as nx
import matplotlib.pyplot as plt
from torch_geometric.data import Data

def load_data_from_csv(csv_path):
    # Load data
    df = pd.read_csv(csv_path)
    
    # Convert categorical labels to integers
    df['label'], _ = pd.factorize(df['label'])
    
    # Normalize numerical features (optional)
    features = df[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']].values
    features = (features - features.mean(axis=0)) / features.std(axis=0)
    
    # Convert DataFrame to node features and edge indices
    num_nodes = len(df)
    num_features = features.shape[1]
    num_classes = len(df['label'].unique())
    
    # Create edge index (fully connected graph for simplicity)
    edge_index = torch.combinations(torch.arange(num_nodes), r=2).t().contiguous()
    
    # Convert to PyTorch tensors
    x = torch.tensor(features, dtype=torch.float)
    y = torch.tensor(df['label'].values, dtype=torch.long)
    
    # Create PyG data object
    data = Data(x=x, edge_index=edge_index, y=y)
    
    return data, num_features, num_classes

def visualize_graph(data, num_nodes_to_show=100):
    # Convert edge index to NetworkX format
    edges = data.edge_index.t().tolist()
    num_nodes = data.x.size(0)
    
    # Sample nodes if the graph is too large
    if num_nodes > num_nodes_to_show:
        sampled_nodes = torch.randperm(num_nodes)[:num_nodes_to_show].tolist()
        sampled_nodes_set = set(sampled_nodes)
        
        # Filter edges to include only those connected to sampled nodes
        filtered_edges = [edge for edge in edges if edge[0] in sampled_nodes_set and edge[1] in sampled_nodes_set]
    else:
        sampled_nodes = list(range(num_nodes))
        filtered_edges = edges
    
    # Create a NetworkX graph
    G = nx.Graph()
    G.add_edges_from(filtered_edges)
    
    # Extract node labels for annotation
    node_labels = {i: f"Node {i}\nFeatures: {data.x[i].tolist()}" for i in sampled_nodes}
    
    # Draw the graph
    pos = nx.spring_layout(G, seed=42)  # Position nodes using spring layout
    plt.figure(figsize=(12, 8))
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, font_weight='bold', edge_color='gray')
    nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=8, verticalalignment='bottom')
    
    plt.title("Graph Visualization")
    plt.show()

# Load data
csv_path = "C:/Users/SRIVATSAL NARAYAN/Desktop/cc/Crop_recommendation.csv"
data, num_features, num_classes = load_data_from_csv(csv_path)

# Visualize the graph
visualize_graph(data)
```
